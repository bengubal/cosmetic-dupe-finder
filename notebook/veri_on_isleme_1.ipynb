```python
import pandas as pd

# ================================
# PRODUCT INFO DOSYASI
# ================================

product_file = "/kaggle/input/datasets/nadyinky/sephora-products-and-skincare-reviews/product_info.csv"

product_columns = [
    "product_id",
    "product_name",
    "brand_name",
    "ingredients",
    "price_usd",
    "rating",
    "reviews",
    "primary_category",
    "secondary_category",
    "tertiary_category"
]

products = pd.read_csv(product_file)

products_selected = products[product_columns].copy()

print("Selected Products Shape:", products_selected.shape)
products_selected.head()
```

```python
review_files = [
    "/kaggle/input/datasets/nadyinky/sephora-products-and-skincare-reviews/reviews_0-250.csv",
    "/kaggle/input/datasets/nadyinky/sephora-products-and-skincare-reviews/reviews_250-500.csv",
    "/kaggle/input/datasets/nadyinky/sephora-products-and-skincare-reviews/reviews_500-750.csv",
    "/kaggle/input/datasets/nadyinky/sephora-products-and-skincare-reviews/reviews_750-1250.csv",
    "/kaggle/input/datasets/nadyinky/sephora-products-and-skincare-reviews/reviews_1250-end.csv"
]

review_columns = [
    "product_id",
    "review_text",
    "rating",
    "is_recommended",
    "helpfulness"
]

review_dfs = []

for file in review_files:
    df = pd.read_csv(file, low_memory=False)
    df_selected = df[review_columns].copy()
    review_dfs.append(df_selected)
    print(file, df_selected.shape)

reviews_selected = pd.concat(review_dfs, ignore_index=True)

print("Combined Reviews Shape:", reviews_selected.shape)
reviews_selected.head()
```

```python
products_selected.to_csv("/kaggle/working/products_selected.csv", index=False)
reviews_selected.to_csv("/kaggle/working/reviews_selected.csv", index=False)
```

```python
# Gerekli kütüphaneler
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

pd.set_option('display.max_colwidth', None)
```

```python
# Ham verisetindeki birincil ürün kategorileri.

print("Primary Category Distribution:")
print(products_selected["primary_category"].value_counts())
```

```python
# Ham verisetindeki ikincil ürün kategorileri

print("Secondary Category Distribution:")
print(products_selected["secondary_category"].value_counts())
```

```python
# Yalnızca içerik benzerliğinin anlamlı olduğu kategoriler alındı.

selected_primary = ["Skincare", "Hair", "Fragrance"]

filtered_products = products_selected[
    products_selected["primary_category"].isin(selected_primary)
].copy()

selected_secondary = [
    "Hair Styling & Treatments",
    "Moisturizers",
    "Treatments",
    "Shampoo & Conditioner",
    "Cleansers",
    "Eye Care",
    "Masks",
    "Sunscreen",
    "Lip Balms & Treatments",
    "Self Tanners"
]

filtered_products = filtered_products[
    filtered_products["secondary_category"].isin(selected_secondary)
].copy()

print("After Category Filtering:", filtered_products.shape)
```

```python
# products verisetindeki eksik değerler analiz edilir

missing_products = (filtered_products.isnull().sum() / len(filtered_products)) * 100
print("Missing Value Percentage in Product Data:")
print(missing_products)
```

```python
# eksik satırlar kaldırılır

filtered_products = filtered_products.dropna(subset=[
    "ingredients",
    "rating",
    "reviews"
]).copy()

print("After Null Value Cleaning:", filtered_products.shape)
```

```python
# Ürün ve marka adlarına göre yinelenen ürünlerin kontrolü

duplicate_count = filtered_products.duplicated(
    subset=["product_name", "brand_name"]
).sum()

print("Duplicate Product Count:", duplicate_count)
```

```python
# ürün ve marka adına göre yinelene ürünler kaldırılır.Örneğin;aynı ürünün hem 30 ml hem de 50 ml versiyonu olabiliyor

filtered_products = filtered_products.drop_duplicates(
    subset=["product_name", "brand_name"]
).copy()

print("After Duplicate Removal:", filtered_products.shape)
```

```python
# Sayısal sütunlarda anormal değerler olup olmadığı incelenir

print(filtered_products[["price_usd", "rating", "reviews"]].describe())
```

```python
# Sayısal aykırı değerlerin incelenmesi için kutu grafiği görselleştirmesi

filtered_products[["price_usd", "reviews"]].boxplot(figsize=(8,4))
plt.title("Boxplot of Price and Review Count")
plt.show()
```

```python
# Aşırı uç değerleri tespit etmek için IQR yöntemi

Q1_price = filtered_products["price_usd"].quantile(0.25)
Q3_price = filtered_products["price_usd"].quantile(0.75)
IQR_price = Q3_price - Q1_price

lower_price = Q1_price - 1.5 * IQR_price
upper_price = Q3_price + 1.5 * IQR_price

Q1_reviews = filtered_products["reviews"].quantile(0.25)
Q3_reviews = filtered_products["reviews"].quantile(0.75)
IQR_reviews = Q3_reviews - Q1_reviews

lower_reviews = Q1_reviews - 1.5 * IQR_reviews
upper_reviews = Q3_reviews + 1.5 * IQR_reviews

print("Price Upper Limit:", upper_price)
print("Reviews Upper Limit:", upper_reviews)
```

```python
# benzerlik analizi için içerik metinleri standartlaştırılır

def clean_ingredients(text):
    text = str(text).lower()
    text = re.sub(r"[\[\]']", "", text)          # remove brackets and quotes
    text = re.sub(r"\([^)]*\)", "", text)        # remove text inside parentheses
    text = re.sub(r"[^a-z,\s]", "", text)        # keep only letters and commas
    text = re.sub(r"\s+", " ", text).strip()     # normalize spaces
    return text

filtered_products["clean_ingredients"] = filtered_products["ingredients"].apply(clean_ingredients)

filtered_products[["product_name", "clean_ingredients"]].head()
```

```python
# Yalnızca temizlenmiş products verisetine ait yorumlar alınır

valid_product_ids = filtered_products["product_id"].unique()

filtered_reviews = reviews_selected[
    reviews_selected["product_id"].isin(valid_product_ids)
].copy()

print("Filtered Reviews Shape:", filtered_reviews.shape)
```

```python
# Dengeli duygu analizi için ürün başına en fazla 40 yorum alınır.

filtered_reviews = filtered_reviews.groupby("product_id").head(40).copy()

print("After Review Sampling:", filtered_reviews.shape)
```

```python
# review verisetindeki eksik değerler analiz edilir

missing_reviews = (filtered_reviews.isnull().sum() / len(filtered_reviews)) * 100
print("Missing Value Percentage in Review Data:")
print(missing_reviews)
```

```python
# Anlamlı review metni içermeyen kayıtlar kaldırılır

filtered_reviews = filtered_reviews.dropna(subset=["review_text"]).copy()

filtered_reviews = filtered_reviews[
    filtered_reviews["review_text"].str.strip() != ""
].copy()

print("After Review Text Cleaning:", filtered_reviews.shape)
```

```python
# Duygu analizi öncesinde yorumlar standartlaştırılır.

def clean_review_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)          # remove URLs
    text = re.sub(r"[^a-z\s]", " ", text)       # remove punctuation and numbers
    text = re.sub(r"\s+", " ", text).strip()    # normalize spaces
    return text

filtered_reviews["clean_review_text"] = filtered_reviews["review_text"].apply(clean_review_text)

filtered_reviews[["review_text", "clean_review_text"]].head()
```

```python
# Anlamsal katkısı zayıf olan aşırı kısa yorumlar kaldırılır.

filtered_reviews = filtered_reviews[
    filtered_reviews["clean_review_text"].str.split().str.len() >= 3
].copy()

print("After Removing Short Reviews:", filtered_reviews.shape)
```

```python
# Save final cleaned datasets for modeling stage

filtered_products.to_csv("/kaggle/working/cleaned_products.csv", index=False)
filtered_reviews.to_csv("/kaggle/working/cleaned_reviews.csv", index=False)

print("Cleaned datasets saved successfully.")
print("Final Product Shape:", filtered_products.shape)
print("Final Review Shape:", filtered_reviews.shape)
```
